```python
from random import *



print('-----------------------------')
print('Welcome to Hangman Game')
print('-----------------------------')
print('Guess the word letter by letter(!!!no character is in caps!!!)')
chance=6

word_list=['automate','audio','webpage','building','player','investment']
num=randint(0,5)
word=word_list[num]
guessed = ['_'] * len(word)

while chance > 0 and '_' in guessed:

    print("Word:", ' '.join(guessed))
    print("Chances left =", chance)

    user_in = input("Enter a letter: ")

    if user_in in word:
        for i in range(len(word)):
            if word[i] == user_in:
                guessed[i] = user_in
    else:
        chance -= 1
        print("Wrong guess!")

if '_' not in guessed:
    print("Congratulations! You won.")
    print("The word was:", word)
else:
    print("Game Over!")
    print("The word was:", word)
```
