```python
def response(message):
    m=message.lower()
    if m=='hello' or m=='hi':
        return 'Hello!!'
    elif m=='how are you':
        return 'I am fine. Thanks!!' 
    elif m=='what is your name':
        return 'My name is chatbot.'
    elif m=='bye':
        return 'Goodbye!'
    else:
        return"Sorry! I don't understand."
def bot():
    print("Bot:Hello! Type 'bye' to exit.")
    while True:
        user=input("You:")
        res=response(user)
        print('Bot:',res)
        if user.lower()=='bye':
            break
start=input("type s to start bot:")
if start.lower()=='s':
    bot()
else:
    pass
```
