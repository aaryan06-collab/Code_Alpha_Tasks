```python
from os import *
import re
import shutil


print('---------------------------------------------')
print('Welcome to Task Automator')
print('---------------------------------------------')

print('1. Move .jpg files to new folder.')
print('2. Copy all email addresses to another file.')

c=int(input('Enter the task number to be performed:'))

if c==1:
    lst=listdir()
    fil_lst=[]
    for i in lst:
        name,ext=path.splitext(i)
        if ext.lower()=='.jpg':
            fil_lst.append(i)
    if len(fil_lst) == 0:
        print("No JPG files found.")
    else:
        chloc=input("enter the new location:")
        if path.exists(chloc):
            for i in fil_lst:
                shutil.move(i, chloc)

            print(len(fil_lst), "jpg files moved successfully.")
        else:
            print("Invalid folder location!")

elif c==2:
    try:
        fname=input('Enter file name(with extension):')
        with open(fname,'r') as fread:
            text=fread.read()
        emails=re.findall(r'[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}', text)
        with open('emails.txt','w') as fwrite:
            for i in emails:
                fwrite.write(i+'\n')
        print("All email addresses are copied successfully!!")

        
    except FileNotFoundError:
        print('File Not Found!!')
else:
    print('Invalid Option')
```
